#### Apply method in pandas

Label encoding is a process of assigning numerical labels to categorical data values. It is a simple and efficient way to convert categorical data into numerical data that can be used for analysis

```python
def posToNum(data):
    d = {'SG':1, 'PF':2, 'PG':3, 'SF':4, 'C':5}
    return d[data]
    
df3['posNum'] = df3.Position.apply(posToNum)
df3
```

```python
# 19 - 25
# 25 - 30
# 30 - 35
# 35 - 40

def ageToCat(data):
    if   19 <= data < 25: return 'A'
    elif 25 <= data < 30: return 'B'
    elif 30 <= data < 35: return 'C'
    elif 35 <= data <=40: return 'D'
    
df3["Age Group"] = df3.Age.apply(ageToCat)
df3
```

### Merging Data in Pandas

```python
l = pd.DataFrame({'id':[1,2,3,4,5], 
                  'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
                  'subject_id':['sub1','sub2','sub4','sub6','sub5']}) 

r = pd.DataFrame({'id':[1,2,3,4,5], 
                  'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
                  'subject_id':['sub2','sub4','sub3','sub6','sub5']}) 
```

```python
l
```

```python
r
```

```python
pd.merge(l, r)
```

```python
pd.merge(l, r, on = 'Name')
```

```python
pd.merge(l, r, on = 'id')      # how = 'inner'  i.e. intersection
```

```python
pd.merge(l, r, on = 'subject_id')      # how = 'inner'  i.e. intersection
```

```python
pd.merge(l, r, on = 'subject_id', how = 'outer')      # outer i.e. union
```

```python
pd.merge(l, r, on = 'subject_id', how = 'left')      # left i.e. outer left join
```

```python
pd.merge(l, r, on = 'subject_id', how = 'right')      # left i.e. outer right join
```

```python
pd.merge(l, r, on = 'subject_id')      # how = 'inner'  i.e. intersection
```

```python
pd.merge(l, r, on = ['id','subject_id'])      # how = 'inner'  i.e. intersection
```
